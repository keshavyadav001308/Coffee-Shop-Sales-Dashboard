```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

username = "root"
password = quote_plus("keshav@123")  # encode special character
host = "localhost"
port = "3306"
database = "starbucks_db"

engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
)

# Push DataFrame to MySQL (this CREATES the table)
df.to_sql(
    name="starbucks_sales",   # table name
    con=engine,
    if_exists="replace",      # creates or replaces table
    index=False
)

print(" Table 'starbucks_sales' created and data inserted!")
```

```python
import pandas as pd

# Convert date
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors='coerce')

# Extract hour safely from time
df['hour'] = df['transaction_time'].apply(lambda x: x.hour)

# Revenue column
df['revenue'] = df['transaction_qty'] * df['unit_price']

# Day name & Month
df['day_name'] = df['transaction_date'].dt.day_name()
df['month'] = df['transaction_date'].dt.month_name()

# Clean column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Drop missing values
df = df.dropna()
```

```python
df.to_excel(
    "coffee_sales_preprocessed.xlsx",
    index=False,
    engine="openpyxl"
)

print(" File saved successfully!")
```
